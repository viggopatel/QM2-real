```python
import rasterio
```

```python
sat19 = rasterio.open('/content/drive/MyDrive/sat2019.tiff')
sat24 = rasterio.open('/content/drive/MyDrive/sat2024.tiff')
```

```python
import rasterio
import numpy as np

# Calculate difference
diff = sat19.read(1) - sat24.read(1)

profile = sat19.profile

with rasterio.open('diff.tif', 'w', **profile) as dst:
    dst.write(diff.astype(rasterio.float32), 1)
```

```python
import matplotlib.pyplot as plt

     # Display the 'diff' data as an image
plt.imshow(diff, cmap='coolwarm')  # You can adjust the colormap as needed
plt.title('Difference Image 2019-2024')
plt.colorbar()  # Add a colorbar to show the data values
plt.show()
```

```python
sat19_data = sat19.read(1)
plt.imshow(sat19_data, cmap='gray')
plt.title('Satellite Image 2019')
plt.show()
```
